```python
import pandas as pd
```

```python
df = pd.read_csv('horse_data.csv')
df.columns = ['Surgery','Age','HospitalNumber','RectalTemperature','Pulse','RespiratoryRate','TemperatureOfExtremities',
              'PeripheralPulse','mucous membranes','capillary refill time','pain','peristalsis','abdominal distension',
             'nasogastric tube', 'nasogastric reflux', 'nasogastric reflux PH', 'rectal examination', 'abdomen',
             'packed cell volume', 'total protein', 'abdominocentesis appearance', 'abdomcentesis total protein',
             'outcome', 'surgical lesion', 'type of lesion1','type of lesion2','type of lesion3', 'cp_data']
df.head(5)
```

```python
df = df.drop(['Age','nasogastric reflux PH','HospitalNumber','PeripheralPulse','TemperatureOfExtremities','peristalsis','nasogastric tube', 'nasogastric reflux','rectal examination', 'abdomen','packed cell volume', 'total protein','abdominocentesis appearance', 'abdomcentesis total protein','outcome', 'surgical lesion', 'type of lesion1','type of lesion2','type of lesion3', 'cp_data'], axis=1)
df.head(15)
```

```python
for col in df.columns:
    df[col] = df[col].apply(lambda x: None if x == '?' else float(x)) 
```

```python
print('мода Операции:', df['Surgery'].mode()[0])
print('мода Слизистой оболочки:', df['mucous membranes'].mode()[0])
print('мода Боли:', df['pain'].mode()[0])
print('мода Вздутия живота:', df['abdominal distension'].mode()[0])
```

Surgery - в исследовании большнство лошадей лечились с опрацией. Данный показатель в выборке наиболее полный

RectalTemperature - большинство лошадей имеют повышенную ректальную температуру, только 25% имеют температуру ниже или равную норме 37.8, максимлаьная очень высокая 40.8, минимальная 35.4

Pulse - большинство лошадей имеют прульс выше нормы 30-40, максимлаьный пульс очень высокий 184

RespiratoryRate - среднее значение частоты дыхания очень высокое 30.43, хотя норма 8-10

mucous membranes - чаще всего у лошадей нормальный розовый цвет слизистой оболочки

capillary refill time - в 75% случаев время наполнения капиляров меньше 3 секунд

pain - чаще всего лошади испытывают периодическую легку боль

abdominal distension - важный показатель болезни, чаще всего лошади не имеют вздутия живота

**Выбросы**

```python
q1 = df.RectalTemperature.quantile(0.25)
q3 = df.RectalTemperature.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
print(f'нижняя гранца выброса: {lower_bound}')
print(f'верхняя гранца выброса: {upper_bound}')
upper_outliers = df[df.RectalTemperature > upper_bound].sort_values('RectalTemperature')
lower_outliers = df[df.RectalTemperature < lower_bound].sort_values('RectalTemperature')
```

```python
print('Выбросы верхней границы:')
upper_outliers
```

```python
print('Выбросы нижней границы:')
lower_outliers
```

Выводы по выбросам ректальной температуры: являются частью исследуемой выборки и не подлежат исключению. Лошади могли иметь в критичиеских ситуациях как очень высокую температуру, так и досточной низкую. Значения выбросов не являются аномалией и имеют место быть.

```python
q1 = df.Pulse.quantile(0.25)
q3 = df.Pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
print(f'нижняя гранца выброса: {lower_bound}')
print(f'верхняя гранца выброса: {upper_bound}')
upper_outliers = df[df.Pulse > upper_bound].sort_values('Pulse')
lower_outliers = df[df.Pulse < lower_bound].sort_values('Pulse')
```

```python
print('Выбросы верхней границы:')
upper_outliers
```

```python
print('Выбросы нижней границы:')
lower_outliers
```

Выводы по выбросам Пульса: выбросов по нижней границе выборки не наблюдается. Выбросы верхней границы находятся в зоне экстремальной нагрузки, но в пределах значений, которые в принципе могут быть у лошадей. Максимальное 250 уд/мин. Исходя из этого выброы оставляем для аналитики и не исключаем.

**Пропуски**

```python
for col in df.columns:
    pct_missing = df[col].isnull().mean()
    print(f'{col} - {pct_missing :.1%}')
```

Решение по замене пропусков: все количественные показатели будем заменять медианой, а качественные модой. Удалять значения не будем, т.к. они имеют существенный процент в выборке и без них данные исказятся и будут недостаточными.
Данный подход позволит не искажать первоначальные данные и не сильно повлияет на базовые показатели.

```python
# Surgery - заменим модой
df.Surgery.fillna(df.Surgery.mode()[0], inplace=True)
# RectalTemperature - заменим медианой
df.RectalTemperature.fillna(df.RectalTemperature.median(), inplace=True)
# Pulse  - заменим медианой
df.Pulse.fillna(df.Pulse.median(), inplace=True)
# RespiratoryRate - заменим медианой
df.RespiratoryRate.fillna(df.RespiratoryRate.median(), inplace=True)
# mucous membranes  - заменим модой
df['mucous membranes'].fillna(df['mucous membranes'].median(), inplace=True)
# capillary refill time  - заменим медианой
df['capillary refill time'].fillna(df['capillary refill time'].mode()[0], inplace=True)
# pain - заменим модой
df['pain'].fillna(df['pain'].mode()[0], inplace=True)
# abdominal distension  - заменим модой
df['abdominal distension'].fillna(df['abdominal distension'].mode()[0], inplace=True)
```
